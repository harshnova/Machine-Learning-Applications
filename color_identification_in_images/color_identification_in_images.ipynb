```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import cv2
from skimage.color import rgb2lab, deltaE_cie76  #deltaE_cie76 is used to calculate the similarity of images
from collections import Counter
import os
import pprint
import PIL.Image
```

```python
def extract_colors(image, max_colors):
    image = cv2.imread(image)
    # the image is somewhat grayed because the original sequence of layers imported by opencv is BGR
    # we therefore need to resequence the layers to RGB
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    # the image is now corrected
    return(getRgbColors(image, max_colors))
```

```python
# the current image is  3 layers RGB intesity map and we need to convert it to hex map
# for that, we need to define this function
def rgbTo2Hex(color):
    return("#{:02x}{:02x}{:02x}".format(int(color[0]), int(color[1]), int(color[2])))
```

This function runs a k-means classifier on the hex values of every pixel obtained into the number of buckets specified by the tester

```python
# extract the colors
def getRgbColors(image, numColors):
    modified_image = cv2.resize(image, (600, 400), interpolation = cv2.INTER_AREA)
    modified_image = modified_image.reshape(modified_image.shape[0]*modified_image.shape[1], 3)
    clf = KMeans(n_clusters = numColors)
    labels = clf.fit_predict(modified_image)
    counts = Counter(labels)
    center_colors = clf.cluster_centers_
    ordered_colors = [center_colors[i] for i in counts.keys()]
    hex_colors = [rgbTo2Hex(ordered_colors[i]) for i in counts.keys()]
    rgb_colors = [ordered_colors[i] for i in counts.keys()]
    plt.figure(figsize = (8, 6))
    plt.pie(counts.values(), labels = hex_colors, colors = hex_colors)      
    return({"hex":hex_colors, "rgb":rgb_colors})
```
